# ÇAPRAZLAMALAR 

```python
import seaborn as sns 
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
cut_kategoriler =  ["Fair","Good","Very Good","Ideal"]
df.cut =df.cut.astype(CategoricalDtype(categories = cut_kategoriler ,ordered = True))
df.head()
```

```python
sns.catplot(x = "cut", y = "price", data = df);
# iki değişkenin birbirlerine göre çaprazlanarak görselleştirmesi gerçekleşmiş oldu.
```

```python
sns.barplot(x = "cut", y = "price",hue = "color",data = df) #hue ile 3.boyutu ekleriz.
# y kısmındaki sayılar değişti.seaborn a üçüncü boyutu ekleyince sütun grafiğinde
# gödtermsei zor old. için bu şekilde verdi.
```

# HİSTOGRAM VE YOĞUNLUK ÇAPRAZLAMALAR

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price,shade = True);
```

```python
(sns
 .FacetGrid(df,            #grafik üzerinde eklenen boyutları göstermek için kullanılır.
              hue ="cut",     #hue boyut ekleme argümanı
              height = 5,     #yükseklik
              xlim = (0,10000))   #grafiğin x ekseni 0 ile 10000 arasında olsun.
.map(sns.kdeplot,"price",shade= True)   #kde plot yoğunluk grafiği oluşturucaz price e göre.facegrid kullanark fonk kullanarak tüm cut sınıflarıyla mapleyeceğiz.
 #bir yoğunluk grafiği oluşturucaz.fakat üzerin cut kategorik değişkeninin sınıflarını boyut olarak eklemek istiyoruz
 # çaprazlayarak değerlendirmek istiyoruz.
.add_legend()        #grafiğin üzerinde eklemiş olduğumuz cut kategorik değişkeninin kategorileri bilgilerini görmek için 
)
```

```python
# standart dünyada bir sayısal değişkenin dağılımını normal dağılım olmasını bekleriz.
# köşelere doğru salmış olmasını bekleriz.fakat burda iki tepeli bir yapı var.
# bu genelde şu anlama gelir odaklanılan sayısal değişkenin içerisindeki bilgiyi yapıyı oluşturan birden fazla 
# faktör var demek olur.
# farklı bir kategorik değişkeni üzerinde ekleyip buna göre bir değerlendirme yapalım dedik.
# bu güzel bir grafik.
# çıkan yorum: elimize bir sayısal değişken geldiğinde farklı kırılımlar ekleyerek de değerlendirmeye çalışmamız gerekir.
```

```python
sns.catplot(x = "cut" , y = "price", hue ="color",kind = "point" ,data = df );
#kind :türü ifade eder point noktalarla ifade edicek.
```

```python
# her farklı renk colorun sınıflarını temsil eder.
# burada cut kategorik değişkeninin ideal sınıfının içerisnde yer alan birbirinden farklı 
# renk olması ve bunların birbirlerinden uzak olması (ideal in dikeye yukarı doğru kısmı)
# şu anlama gelir color aracılığıyla cut değişkeni sınıfları kesişiminde bir bilgi taşındığı
# birbirinden farklı bir bilgi taşındığı anlamına geliyor.ama mesela burda mor ve mavi için geçerli değil.
# cut un ideal sınıfı içerisinde bu iki sınıfının taşıdığı bilgi aynı seviyeleri aynı.diğer renklilerde farklılıaşmış
# 3 değişkenli buradaki gibi bir veri seti olsun :price,color,cut bunu makine öğrenmesi dediğimiz yöntemlerle
# modellemeye başladığımızda  arka taraftaki algoritmaların en büyük amacı price değişkeninin içindeki bilgiyi 
# bu değişimi(3000-6000)  diğer değişkenlerce açıklamaya çalışmak oluyor.
# diğer değişkenlerde price ın açıklanma çabasının en önemli noktası ise  price içinde bu değişkenlerin ayırtedici 
# bilgi taşımsı oluyor.
```
